```python
def searchMatrix(matrix,target):
    M=len(matrix)
    N=len(matrix[0])

    point1i=0
    point1j=0
    point2i=M-1
    point2j=N-1

    if(target<matrix[point1i][point1j] or target>matrix[point2i][point2j]):
        return False

    while point1i!=point2i or point1j!=point2j:
        if(matrix[point1i][point1j]==target):
            return True
        if(point1i==point2i):
            if(matrix[point1i][point1j]<target):
                point1j+=1
                continue
            else:
                point2j-=1
                continue
        else:
            if(matrix[point1i+1][point1j]<=target):
                point1i+=1
                continue
            else:
                point2i-=1
                continue
    if(matrix[point1i][point1j]!=target):
        return False
    else:
        return True
searchMatrix([[1],[3]],3)
```

```python
nums = [1,2,3]
N = len(nums)
res=[]
def digui(cur,idx):
    if(cur not in res):
        res.append(cur)
    for i in range(idx,N):
        digui(cur+[nums[i]],i+1)
digui([],0)
res
```

```python
board = [["A","B","C","E"],
         ["S","F","C","S"],
         ["A","D","E","E"]]
word = "ABCCED"
M = len(board)
N = len(board[0])
res=[]
path=[]
def digui(i,j,path,count):
    if(count==len(word) and path not in res):
        res.append(path)
        return True
    if(board[i][j]==word[count] and [i,j] not in path):
        try:
            digui(i+1,j,path+[[i,j]],count+1)
            digui(i-1,j,path+[[i,j]],count+1)
            digui(i,j+1,path+[[i,j]],count+1)
            digui(i,j-1,path+[[i,j]],count+1)
        except:
            return False
for i in range(M):
    for j in range(N):
        if board[i][j]==word[0]:
            digui(i,j,path,0)
print(res)
if res:
    print('True')
else:
    print('False')
```
